5.Write a python code to display the following pie chart

```python
plt.title('A Pie Chart',fontsize=15,fontname='Times New Roman')
y = np.array([7.143,21.429,39.286,10.714,21.429])
mylabels = ["CSE","ECE","Mech","Civil","BioTech"]
myexplode = [0.9, 0.2, 0, 0, 0]
colors = ['pink','magenta','yellow','blue','red']
plt.pie(y, labels = mylabels,colors=colors ,explode = myexplode,autopct=
        '%1.3f%%',startangle=180)
plt.show()
```
